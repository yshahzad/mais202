```python
import pandas as pd

weighthistory_df = pd.read_sas("WeightHistory_2017_18.xpt")
foodsecurity_df = pd.read_sas("FoodSecurity_2017_18.xpt")
smokinghist_df = pd.read_sas("SmokingHistory_CigUse_2017_18.xpt")
secondhandsmoke_df = pd.read_sas("SmokingHistorySecondHand_2017_18.xpt")
rxdrugs_df = pd.read_sas("PrescriptionDrugs_2017_18.xpt")
druguse_df = pd.read_sas("DrugUse_2017_18.xpt")
nicotinemetabolites = pd.read_sas("NicotineMetabolites_2017_18.xpt")
hdl_df = pd.read_sas("CholesterolHDL_2017_18.xpt")
ldl_trig_df = pd.read_sas("CholesterolLDL_2017_18.xpt")


nicotinebysmoke_df = smokinghist_df.merge(right=nicotinemetabolites, how="inner", on="SEQN")

print(nicotinebysmoke_df)
```

Notes:

Questionnaire data from : https://wwwn.cdc.gov/nchs/nhanes/search/datapage.aspx?Component=Questionnaire&CycleBeginYear=2017

*Nicotine by Smoking History variable names:*

Questionnaire:
* SMD030 - Age started smoking cigarettes regularly
* SMD641 - # days smoked cigs during past 30 days
* SMQ078 - How soon after waking do you smoke
* SMQ050Q - How long since quit smoking cigarettes

*Lab*
* URXCOTT - Cotinine, urine (ng/mL)
* URXNNCT - Nornicotine, urine (ng/mL)
